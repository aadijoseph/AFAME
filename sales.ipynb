```python
df["Total_Sales"] = df["Sales"] * df["Quantity"]
```

```python
df
```

```python
sum(df["Total_Sales"])
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
# Extract year from the 'Date' column
df['Year'] = df['Order Date'].dt.year
```

```python
df
```

```python
grouped_data = df.groupby(['Category', 'Year']).sum()['Total_Sales'].unstack()


# Plot the trend for each category
plt.figure(figsize=(10, 6))
for category in grouped_data.columns:
    plt.plot(grouped_data.index, grouped_data[category], label=category)

plt.title('Yearly Sales Trend by Category')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
grouped_data = df.groupby(['Sub-Category', 'Year']).sum()['Total_Sales'].unstack()


# Plot the trend for each category
plt.figure(figsize=(10, 6))
for category in grouped_data.columns:
    plt.plot(grouped_data.index, grouped_data[category], label=category)

plt.title('Yearly Sales Trend by Category')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Product Category wise Total Sales
plt.figure(figsize=(10,6))
sns.barplot(x="Sub-Category",y="Total_Sales",data = df)
plt.title = ("Product Category wise Total Sales")
plt.xlabel = ("Sub-Category")
plt.ylabel = ("Sales")
plt.xticks(rotation=45, ha='right')
plt.tight_layout(pad=2.0)
plt.show()
```

```python
# Plot box plot of sales by region
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Total_Sales', data=df)
plt.title('Sales Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout(pad=2.0)
plt.show()
```

```python
# Plot the histogram using Seaborn
plt.figure(figsize=(8, 6))
sns.histplot(df, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Data (Seaborn)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
